```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://pwonlyias.com/gdp-of-indian-states/#:~:text=Maharashtra%20holds%20the%20top%20position,the%20fiscal%20year%202023%2D24"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup
```

```python
table = soup.find("table")
```

```python
table.find_all('tr')
```

```python
data = []
headers = []
```

```python
for th in table.find_all("th"):
    headers.append(th.text.strip())
```

```python
for tr in table.find_all("tr"):
    row = []
    for td in tr.find_all("td"):
        row.append(td.text.strip())
    if row:
        data.append(row)
```

```python
data
```

```python
headers
```

```python
df = pd.DataFrame(data, columns=headers)
```

```python
df
```

```python
df.to_csv('State GDP Analysis.csv')
```

```python
import numpy as np
projected_gdp = np.array([14.49, 0.37, 5.67, 8.59, 5.07, 10.4, 1, 25.62, 11.2, 2.14,
                          2.3, 4.23, 25, 11.3, 13.87, 38.79, 0.45, 0.46, 0.35, 0.37,
                          8.65, 6.98, 15.7, 0.42, 14, 0.26, 24.39, 3.33, 17.19])

current_gdp = np.array([2.19, 2.05, 1.18, 0.54, 1.33, 2.72, 4.72, 2.41, 2.96, 2.22,
                        1.32, 0.91, 3.01, 2.33, 1.4, 2.24, 0.91, 0.98, 1.98, 1.25,
                        1.5, 1.73, 1.56, 5.19, 3.08, 1.59, 0.83, 2.33, 1.41])

median_projected_gdp = np.median(projected_gdp)
median_current_gdp = np.median(current_gdp)

# Print the average values
print(f"Median Projected GDP: {median_projected_gdp:} Rs Lakh Crore")
print(f"Median Current GDP: {median_current_gdp:} Rs Lakh")
```

```python
max_projected_gdp = np.max(projected_gdp)
min_projected_gdp = np.min(projected_gdp)

max_current_gdp = np.max(current_gdp)
min_current_gdp = np.min(current_gdp)

states = ["Andhra Pradesh", "Arunachal Pradesh", "Assam", "Bihar", "Chhattisgarh",
          "Delhi", "Goa", "Gujarat", "Haryana", "Himachal Pradesh",
          "Jammu & Kashmir (UT)", "Jharkhand", "Karnataka", "Kerala",
          "Madhya Pradesh", "Maharashtra", "Manipur", "Meghalaya", "Mizoram",
          "Nagaland", "Odisha", "Punjab", "Rajasthan", "Sikkim", "Telangana",
          "Tripura", "Uttar Pradesh", "Uttarakhand", "West Bengal"]

max_projected_state = states[np.argmax(projected_gdp)]
min_projected_state = states[np.argmin(projected_gdp)]

max_current_state = states[np.argmax(current_gdp)]
min_current_state = states[np.argmin(current_gdp)]

print(f"Max Projected GDP: {max_projected_gdp:} Rs Lakh Crore (State: {max_projected_state})")
print(f"Min Projected GDP: {min_projected_gdp:} Rs Lakh Crore (State: {min_projected_state})")

print(f"Max Current GDP: {max_current_gdp:} Rs Lakh (State: {max_current_state})")
print(f"Min Current GDP: {min_current_gdp:} Rs Lakh (State: {min_current_state})")
```

```python
growth_rates = ((projected_gdp - current_gdp) / current_gdp) * 100

max_growth_rate = np.max(growth_rates)
min_growth_rate = np.min(growth_rates)

max_growth_rate_state = states[np.argmax(growth_rates)]
min_growth_rate_state = states[np.argmin(growth_rates)]

print(f"Max Growth Rate: {max_growth_rate:}% ({max_growth_rate_state})")
print(f"Min Growth Rate: {min_growth_rate:}% ({min_growth_rate_state})")
```
